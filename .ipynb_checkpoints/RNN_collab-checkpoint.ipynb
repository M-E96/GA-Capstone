```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, GRU
from tensorflow.keras.utils import timeseries_dataset_from_array
```

```python
temp = pd.read_csv("/content/drive/MyDrive/Capstone Data/temperature-anomaly.csv")
```

```python
temp.head()
```

```python
temp = temp[["Entity", "Year", "Global average temperature anomaly relative to 1961-1990"]]
```

```python
temp.head()
```

```python
temp = temp[temp["Entity"] == "Global"]
```

```python
temp = temp.iloc[110:-1]
```

```python
temp.head()
```

```python
temp.drop(columns= "Entity", inplace= True)
temp.head()
```

```python
temp.tail(200)
```

```python
temp.Year.shape
```

```python
X = temp.Year
y = temp["Global average temperature anomaly relative to 1961-1990"]

X = np.array(X)
y = np.array(y)

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, shuffle= False)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train.shape
```

```python
seq_len = 1
batch_size = 16


train_gen = timeseries_dataset_from_array(
    X_train_sc, y_train,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)

val_gen = timeseries_dataset_from_array(
    X_test_sc, y_test,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)
```

```python
model_1 = Sequential([
    LSTM(64, return_sequences= True, input_shape= (seq_len, 1)),
    LSTM(32),
    Dense(1, activation= "linear")
])
```

```python
model_1.compile(loss= "mse", optimizer= "adam", metrics= ["mae"])
```

```python
hist = model_1.fit(
    train_gen,
    validation_data= val_gen,
    epochs= 20
)
```

```python
plt.plot(hist.history["loss"], c= "crimson", label= "Train")
plt.plot(hist.history["val_loss"], c= "navy", label= "Val")
plt.legend()
```

```python
pred_1 = model_1.predict(X_test_sc)
```

```python
plt.plot(y_test, color= "navy", label= "y_test")
plt.plot(pred_1, color= "crimson", label= "pred")
plt.legend()
plt.title("Time correlation")
```

```python
mse = mean_squared_error(y_test, pred_1)
mse
```

```python
np.sqrt(mse)
```

RNN best used for regression/verification than predicting based on this data.
I tried bigger sequence lengths, but model deteriorates beyond length 1, so future forecasting is out of the picture really.
After 1, all predictions converged to 0

```python
X.shape, y.shape
```

```python
X = np.array(X)
y = np.array(y)

X = np.array(X)
y = np.array(y)

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, shuffle= False)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
y.shape
```

```python
seq_len = 1
batch_size = 16


train_gen = timeseries_dataset_from_array(
    X_train_sc, y_train,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)

val_gen = timeseries_dataset_from_array(
    X_test_sc, y_test,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)
```

```python
pred_2 = model_2.predict(X_test_sc)
plt.plot(y_test, color= "navy", label= "y_test")
plt.plot(pred_2, color= "crimson", label= "pred")
plt.legend()
plt.title("CO2 correlation")
```

```python
mse = mean_squared_error(y_test, pred_2)
mse, np.sqrt(mse)
```

Having gone through different iterations of the RNN, they both ended up preferring the same model, and the CO2 seemed to have a better mse/rmse than just time

Now we work on predicting disasters

```python
X = df[["Annual CO2 emissions", "Global average temperature anomaly relative to 1961-1990", "Average sea level"]]
y = df["Disasters"]
```

```python
X.head()
```

```python
y.head()
```

```python
X.tail()
```

```python
y.tail()
```

```python
X.shape, y.shape
```

```python
X = np.array(X)
y = np.array(y)

X = np.array(X)
y = np.array(y)

y = y.reshape(-1, 1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, shuffle= False)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X.shape, y.shape
```

```python
seq_len = 1
batch_size = 32


train_gen = timeseries_dataset_from_array(
    X_train_sc, y_train,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)

val_gen = timeseries_dataset_from_array(
    X_test_sc, y_test,
    sequence_length = seq_len,
    sampling_rate= 1,
    batch_size= batch_size
)
```

```python
model_3 = Sequential([
    LSTM(128, return_sequences= True, input_shape= (seq_len, 3)),
    LSTM(164, return_sequences= True),
    LSTM(64),
    Dense(64, activation= "relu"),
    Dense(32, activation= "relu"),
    Dense(1, activation= "linear")
])
```

```python
model_3.compile(loss= "mse", optimizer= "adam", metrics= ["mae"])
hist3 = model_3.fit(
    train_gen,
    validation_data= val_gen,
    epochs= 60
)
```

```python
plt.plot(hist3.history["loss"], c= "crimson", label= "Train")
plt.plot(hist3.history["val_loss"], c= "navy", label= "Val")
plt.legend()
```

```python
X_test_sc_reshaped = X_test_sc.reshape((X_test_sc.shape[0], 1, X_test_sc.shape[1]))
pred_3 = model_3.predict(X_test_sc_reshaped)
plt.plot(y_test, color= "navy", label= "y_test")
plt.plot(pred_3, color= "crimson", label= "pred")

plt.legend()
plt.title("Disaster Correlation")
```

```python
mse = mean_squared_error(y_test, pred_3)
mse, np.sqrt(mse)
```

This isn't the best prediction, but considering what the data looked like, it's understandable it turned out like this.
